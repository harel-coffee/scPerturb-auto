## manually split sciplex3 by cell line and prepare pca for these

```python
key = 'SrivatsanTrapnell2020_sciplex3'
```

```python
# split by cell_line and process separately
path = h5_files[key]
alldata = sc.read(path)
for cl in tqdm(['A549', 'K562', 'MCF7']):
    adata = alldata[alldata.obs.cell_line==cl].copy()
    adata.layers['counts'] = adata.X.copy()
    # basic qc and pp
    sc.pp.filter_cells(adata, min_counts=1000)
    sc.pp.normalize_per_cell(adata)
    sc.pp.filter_genes(adata, min_cells=50)
    sc.pp.log1p(adata)

    # high class imbalance
    adata = equal_subsampling(adata, 'perturbation', N_min=50)
    sc.pp.filter_genes(adata, min_cells=3)  # sanity cleaning

    # select HVGs
    n_var_max = 2000  # max total features to select
    sc.pp.highly_variable_genes(adata, n_top_genes=n_var_max, subset=False, flavor='seurat_v3', layer='counts')
    sc.pp.pca(adata, use_highly_variable=True)
    sc.pp.neighbors(adata)
    adata.write(SDIR+f'tmp_data_{key}_{cl}.h5')
```

## compute E-distances

```python
# Selection of datasets where looking at the E-distance is meaningful
# (enough perturbations, not too many multi-perturbations, well-defined control)
datasets = ['TianKampmann2021_CRISPRa',
 'TianKampmann2019_day7neuron',
 'FrangiehIzar2021_RNA',
 'DixitRegev2016',
 'PapalexiSatija2021_eccite_RNA',
 'DatlingerBock2017',
 'DatlingerBock2021',
 'ReplogleWeissman2022_rpe1',
 'SchraivogelSteinmetz2020_TAP_SCREEN__chromosome_11_screen',
 'TianKampmann2019_iPSC',
 'TianKampmann2021_CRISPRi',
 'ReplogleWeissman2022_K562_essential',
 'PapalexiSatija2021_eccite_arrayed_RNA',
 'NormanWeissman2019_filtered'] + \
[
 'SrivatsanTrapnell2020_sciplex3_A549',
 'SrivatsanTrapnell2020_sciplex3_K562',
 'SrivatsanTrapnell2020_sciplex3_MCF7'
]
```

```python
# set colors
colors_perturbation_types = {
    'CRISPRi': 'tab:blue', 
    'CRISPRa': 'tab:red', 
    'CRISPR': 'tab:orange',
    'CRISPR-cas9': 'tab:orange',
    'drugs': 'tab:green',
    'drug': 'tab:green',
    'cytokine': 'tab:olive',
    'cytokines': 'tab:olive',
    'drugs, CRISPR-cas9': 'tab:pink',
    'CRISPR-cas9, TCR stim': 'tab:purple'
}
```

```python
# Compute E-distances with pairwise pca distances pre-computed with Snakemake
estats_datasets = {}
estats_control_datasets = {}
for key in tqdm(datasets):
    if not os.path.exists(f'{table_path}/pairwise_pca_distances_{key}_tables.csv'):
        continue
    # load averaged pca pairwise distances
    pwd = pd.read_csv(f'{table_path}/pairwise_pca_distances_{key}_tables.csv', index_col=0)
    if 'control' not in pwd.index or len(pwd.index)<3:
        continue
    
    # derive basic statistics
    sigmas = np.diag(pwd)
    deltas = pwd
    
    # compute E-distances
    estats = 2 * deltas - sigmas - sigmas[:, np.newaxis]  # this is matrix notation
    estats = cluster_matrix(estats, 'both')  # cluster resulting matrix (hierarchically)
    estats_control = estats.loc[:, 'control']
    
    # record results
    estats_control_datasets[key] = estats_control
    estats_datasets[key] = estats
```

```python
# aggregate
df = pd.concat(estats_control_datasets).reset_index()
df.columns = ['dataset', 'perturbation', 'E-distance']
# prettier names
df['dataset_pretty'] = [x.replace('TAP_SCREEN__', '') for x in df.dataset]
df['dataset_pretty'] = [x.replace('_', '\n(', 1)+')' if '_' in x else x for x in df.dataset_pretty]
df['dataset_pretty'] = [x.replace('_', ' ') for x in df.dataset_pretty]
```

```python
# This is the newest excel table with the dataset metadata. We extract the perturbation_type.
# Note that this is only for datasets available at time of publication!
metadata = pd.read_csv('../metadata/scperturb_dataset_info_140922.csv', index_col=0)
metadata['dataset_id'] = [i if pd.isna(x) else f'{i}_{x}' for i, x in zip(metadata.index, metadata.dataset_index)]
metadata = metadata.set_index('dataset_id')
metadata.head()
```

```python
# add more detailed perturbation_type annotation
df['perturbation_type_'] = None
for dataset in pd.unique(df.dataset):
    dataset_ = 'SrivatsanTrapnell2020_sciplex3' if 'SrivatsanTrapnell2020_sciplex3' in dataset else dataset
    val = metadata.loc[dataset_, 'Perturbation'].replace(', TCR stim', '')
    df.loc[df.dataset==dataset, 'perturbation_type_'] = val
```
